```sql
with state_emissions as (
  select
    state_abbr,
    sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double)) as total_emissions
  from emissions.default.emissions_data
  group by state_abbr
  order by total_emissions desc
  limit 10
)
select
  sum(total_emissions) as total_top10_emissions,
  (sum(total_emissions) /  (select(sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double))) from emissions_data)) * 100 as percentage_top10_emissions
from state_emissions
```
